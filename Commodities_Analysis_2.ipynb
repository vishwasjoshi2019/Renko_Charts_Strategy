# Month wise Correlation Calculation for last 20 years for Wheat & Corn

```python
import yfinance as yf
import pandas as pd

tickers = ['KE=F', 'ZC=F']

# Define the time frame
start_date = '2004-01-01'  # Start from January 2004
end_date = '2023-11-30'
time_frame = '1d'

# Fetch historical data for each ticker
data1 = yf.download(tickers, start=start_date, end=end_date, interval=time_frame)['Adj Close']

# Create an empty DataFrame to store correlation values
correlation_df1 = pd.DataFrame()

# Calculate the correlation between the two commodities for each month
for year in range(int(start_date[:4]), int(end_date[:4]) + 1):
    for month in range(1, 13):
        date_str = f"{year}-{month:02d}"
        if date_str in data1.index:
            month_data = data1[date_str]
            correlation_value = month_data.pct_change().corr().iloc[0, 1]  # Extract the correlation value
            correlation_df1.loc[date_str, 'Correlation'] = correlation_value

# Display the resulting DataFrame
print(correlation_df1)
```

# Week wise daily returns of Wheat and Corn for last 5 years

```python
import yfinance as yf
import pandas as pd

tickers = ['KE=F', 'ZC=F']

# Define the time frame
start_date2 = '2019-01-01'  # Start from January 2019
end_date2 = '2023-11-30'
time_frame = '1d'  # Daily data

# Fetch historical data for each ticker
data = yf.download(tickers, start=start_date2, end=end_date2, interval=time_frame)['Adj Close']

# Calculate daily percentage change
daily_returns = data.pct_change()

weekly_returns = daily_returns.resample('W-Fri').mean()

weekly_returns_df1 = pd.DataFrame({
    'Date': weekly_returns.index,
    'KE=F_Return': weekly_returns['KE=F'].values,
    'ZC=F_Return': weekly_returns['ZC=F'].values
})
weekly_returns_df1
```

```python
import yfinance as yf
import pandas as pd
!pip install ta
from ta.trend import EMAIndicator

tickers = ['KE=F', 'ZC=F']

# Define the time frame
start_date = '2004-01-01'  # Start from January 2004
end_date = '2023-11-30'
time_frame = '1d'

# Fetch historical data for each ticker
data = yf.download(tickers, start=start_date, end=end_date, interval=time_frame)['Adj Close']

# Create an empty DataFrame to store correlation and EMA values
result_df = pd.DataFrame()

# Calculate the correlation between the two commodities and EMAs for each month
for year in range(int(start_date[:4]), int(end_date[:4]) + 1):
    for month in range(1, 13):
        date_str = f"{year}-{month:02d}"
        if date_str in data.index:
            month_data = data[date_str]
            correlation_value = month_data.pct_change().corr().iloc[0, 1]  # Extract the correlation value

            # Calculate EMAs
            ema_3 = EMAIndicator(close=month_data, window=3).ema_indicator()
            ema_6 = EMAIndicator(close=month_data, window=6).ema_indicator()
            ema_12 = EMAIndicator(close=month_data, window=12).ema_indicator()

            # Append data to the DataFrame
            result_df.loc[date_str, 'Correlation'] = correlation_value
            result_df.loc[date_str, 'EMA_3'] = ema_3.iloc[-1] if not ema_3.empty else None
            result_df.loc[date_str, 'EMA_6'] = ema_6.iloc[-1] if not ema_6.empty else None
            result_df.loc[date_str, 'EMA_12'] = ema_12.iloc[-1] if not ema_12.empty else None


# Display the resulting DataFrame
print(result_df)
```
