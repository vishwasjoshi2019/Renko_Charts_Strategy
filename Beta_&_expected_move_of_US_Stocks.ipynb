# **Dow Jones Top 20 Stocks Beta**

```python
import yfinance as yf
import pandas as pd
import numpy as np

# Define a dictionary to map symbols to names
stock_names = {
    "UNH": "Unitedhealth Group Inc",
    "GS": "Goldman Sachs Group Inc",
    "MSFT": "Microsoft Corp",
    "HD": "Home Depot Inc",
    "CAT": "Caterpillar Inc",
    "AMGN": "Amgen Inc",
    "MCD": "Mcdonald S Corp",
    "V": "Visa Inc Class a Shares",
    "CRM": "Salesforce Inc",
    "BA": "Boeing Co",
    "HON": "Honeywell International Inc",
    "CVX": "Chevron Corp",
    "AAPL": "Apple Inc",
    "TRV": "Travelers Cos Inc",
    "WMT": "Walmart Inc",
    "JNJ": "Johnson & Johnson W/d",
    "AXP": "American Express Co",
    "PG": "Procter & Gamble Co",
    "JPM": "Jpmorgan Chase & Co",
    "IBM": "Intl Business Machines Corp"
}

# Define the list of stock symbols
stock_symbols = list(stock_names.keys())

# Download historical data for DJIA
djia = yf.download("^DJI", start="2023-01-01", end="2023-12-31", progress=False)['Adj Close']

# Create a DataFrame to store beta values
beta_df = pd.DataFrame(columns=["Company", "Symbol", "Beta"])

# Calculate beta for each stock
for symbol in stock_symbols:
    # Download historical data for the stock
    stock_data = yf.download(symbol, start="2023-01-01", end="2023-12-31", progress=False)['Adj Close']

    # Check if the date ranges match
    if not djia.index.equals(stock_data.index):
        print(f"Date ranges for {stock_names[symbol]} and DJIA do not match. Skipping.")
        continue

    # Calculate daily returns for both DJIA and the stock
    djia_returns = djia.pct_change().dropna()
    stock_returns = stock_data.pct_change().dropna()

    # Calculate beta using the covariance and variance
    covariance = np.cov(stock_returns, djia_returns)
    stock_variance = covariance[0, 0]
    market_variance = covariance[1, 1]
    beta = round(covariance[0, 1] / market_variance, 3)

    # Append the result to the DataFrame with actual company names
    beta_df = beta_df.append({"Company": stock_names[symbol], "Symbol": symbol, "Beta": beta}, ignore_index=True)

# Print the DataFrame with beta values including company names
beta_df
```

# **DOW JONES TOP 20 STOCKS EXPECTED MOVE USING BETA**

```python
import yfinance as yf
import pandas as pd
import numpy as np

# Define a dictionary to map symbols to names
stock_names = {
    "UNH": "Unitedhealth Group Inc",
    "GS": "Goldman Sachs Group Inc",
    "MSFT": "Microsoft Corp",
    "HD": "Home Depot Inc",
    "CAT": "Caterpillar Inc",
    "AMGN": "Amgen Inc",
    "MCD": "Mcdonald S Corp",
    "V": "Visa Inc Class a Shares",
    "CRM": "Salesforce Inc",
    "BA": "Boeing Co",
    "HON": "Honeywell International Inc",
    "CVX": "Chevron Corp",
    "AAPL": "Apple Inc",
    "TRV": "Travelers Cos Inc",
    "WMT": "Walmart Inc",
    "JNJ": "Johnson & Johnson W/d",
    "AXP": "American Express Co",
    "PG": "Procter & Gamble Co",
    "JPM": "Jpmorgan Chase & Co",
    "IBM": "Intl Business Machines Corp"
}

# Define the list of stock symbols
stock_symbols = list(stock_names.keys())

# Download historical data for DJIA
djia_data = yf.download("^DJI", start="2023-01-01", end="2023-12-31", progress=False)
djia_returns = djia_data["Adj Close"].pct_change().dropna()

# Calculate the average percentage move for DJIA in different timeframes
djia_1_day_avg_move = djia_returns.mean()
djia_5_day_avg_move = djia_returns.rolling(window=5).mean().iloc[-1]
djia_15_day_avg_move = djia_returns.rolling(window=15).mean().iloc[-1]

# Create a DataFrame to store beta values and expected moves
beta_df = pd.DataFrame(columns=["Company", "Symbol", "Beta", "1-Day DJIA Avg Move", "5-Day DJIA Avg Move", "15-Day DJIA Avg Move"])

# Calculate beta and expected moves for each stock
for symbol in stock_symbols:
    # Download historical data for the stock
    stock_data = yf.download(symbol, start="2023-01-01", end="2023-12-31", progress=False)

    # Check if the date ranges match
    if not djia_data.index.equals(stock_data.index):
        print(f"Date ranges for {stock_names[symbol]} and DJIA do not match. Skipping.")
        continue

    stock_returns = stock_data["Adj Close"].pct_change().dropna()

    # Calculate beta using the covariance and variance of daily returns
    covariance = np.cov(stock_returns, djia_returns)
    stock_variance = covariance[0, 0]
    market_variance = covariance[1, 1]
    beta = round(covariance[0, 1] / market_variance, 3)

    # Calculate the expected moves for the stock using beta and DJIA average moves in percentage
    stock_1_day_expected_move = beta * djia_1_day_avg_move
    stock_5_day_expected_move = beta * djia_5_day_avg_move
    stock_15_day_expected_move = beta * djia_15_day_avg_move

    # Append the results to the DataFrame
    beta_df = beta_df.append({"Company": stock_names[symbol], "Symbol": symbol, "Beta": beta,
                              "1-Day DJIA Avg Move": djia_1_day_avg_move,
                              "5-Day DJIA Avg Move": djia_5_day_avg_move,
                              "15-Day DJIA Avg Move": djia_15_day_avg_move,
                              "1-Day Expected Move": stock_1_day_expected_move,
                              "5-Day Expected Move": stock_5_day_expected_move,
                              "15-Day Expected Move": stock_15_day_expected_move}, ignore_index=True)

# Print the DataFrame with beta values and expected moves
beta_df
```
